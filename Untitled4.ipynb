## Data loading

### Subtask:
Load the "Bihar Air Quality Analysis.csv" dataset.


**Reasoning**:
Load the dataset and display its first few rows and shape to confirm successful loading.



```python
import pandas as pd

try:
    df = pd.read_csv('Bihar Air Quality Analysis.csv')
    display(df.head())
    print(df.shape)
except FileNotFoundError:
    print("Error: 'Bihar Air Quality Analysis.csv' not found.")
except pd.errors.ParserError:
    print("Error: Unable to parse the CSV file. Please check its format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

## Data exploration

### Subtask:
Explore the dataset to understand its structure, identify data types, check for missing values, and examine the distribution of key variables.


**Reasoning**:
Examine the data types, missing values, descriptive statistics, time period, outliers and column names.



```python
# Data Type Identification
print("Data Types:\n", df.dtypes)

# Missing Value Analysis
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
print("\nMissing Values:\n", missing_values)
print("\nMissing Value Percentage:\n", missing_percentage)

# Descriptive Statistics
numerical_cols = df.select_dtypes(include=['number']).columns
print("\nDescriptive Statistics:\n", df[numerical_cols].describe())

# Time Period Analysis (Not applicable as there's no date/time column)
print("\nNo date/time column found.")

# Outlier Detection (Visual inspection of numerical columns)
# Due to lack of air quality parameters, we'll only check for outliers in 'Sum of doctorCount'
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.boxplot(df['Sum of doctorCount'], vert=False)
plt.title('Boxplot of Sum of doctorCount')
plt.show()

# Column Name Examination
print("\nColumn Names:", df.columns.tolist())
print("\nThe data does not seem to be related to air quality.")
```

## Data cleaning

### Subtask:
Clean the data by handling missing values and renaming columns.


**Reasoning**:
Drop unnecessary columns, impute missing values in 'Sum of doctorCount', rename columns for clarity, and verify the absence of missing values.



```python
# Drop unnecessary columns
df = df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5'])

# Impute missing values in 'Sum of doctorCount' using the mean
mean_doctor_count = df['Sum of doctorCount'].mean()
df['Sum of doctorCount'] = df['Sum of doctorCount'].fillna(mean_doctor_count)

# Rename columns
df = df.rename(columns={'Row Labels': 'State/District', 'Sum of doctorCount': 'Total Doctors'})

# Verify no missing values
print(df.isnull().sum())
display(df)
```

**Reasoning**:
The previous code block still has a missing value in 'State/District'. I will impute that value and then verify again.



```python
# Impute the missing value in 'State/District'
df['State/District'] = df['State/District'].fillna('Unknown')

# Verify no missing values
print(df.isnull().sum())
display(df)
```

## Data visualization

Generate the visualizations (bar chart, scatter plot, box plot) and document the observations as per the instructions.
