## Data cleaning

### Subtask:
Clean the data by handling missing values and outliers in the 'Sum of doctorCount' column.


**Reasoning**:
Impute the missing value in 'Sum of doctorCount' with the median, handle outliers by winsorizing at the 99th percentile, and check for inconsistencies like negative values.



```python
# Impute missing value in 'Sum of doctorCount' with the median
# Rationale: The median is less sensitive to outliers than the mean.
df['Sum of doctorCount'] = df['Sum of doctorCount'].fillna(df['Sum of doctorCount'].median())

# Handle outliers by winsorizing at the 99th percentile
# Rationale: Winsorizing caps extreme values, reducing their influence on analysis without removing data points.
upper_limit = df['Sum of doctorCount'].quantile(0.99)
df['Sum of doctorCount'] = df['Sum of doctorCount'].clip(upper=upper_limit)

# Check for inconsistencies (negative values)
# Rationale: Negative doctor counts are not meaningful, so any negative values after winsorizing must be addressed.
if (df['Sum of doctorCount'] < 0).any():
  print("Negative values found in 'Sum of doctorCount' after outlier handling.")
  # Replace negative values with 0
  df['Sum of doctorCount'] = df['Sum of doctorCount'].clip(lower=0)
else:
  print("No negative values found in 'Sum of doctorCount'.")

display(df)
```
